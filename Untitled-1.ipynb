```python
categories = soup.find_all('div',class_= "card-title")  # Update CSS selector
for category in categories:
        category_name = category.text.strip()
        print(category_name)  
```

```python

subcategory_links = soup.select("div.mdc-list.dataset-list a.mdc-list-item")
base_url = "https://wesis.org"
target_indices = [3]

print(f"Total links found: {len(subcategory_links)}")

for index, link in enumerate(subcategory_links):
    if index not in target_indices:
        continue
        
    print(f"\nProcessing index {index}")
    
    
    href = link.get("href")
    print(f"Extracted href: {href}")  # Debug print
    
    if not href:
        print(f"No href found for index {index}")
        continue
        
    text = link.get_text(strip=True)
    full_url = base_url + href
    print(f"Full URL: {full_url}")  # Debug print
    
```

```python
driver.get(full_url)

subcategory_name = driver.find_element(By.CLASS_NAME, "title").text
wait = WebDriverWait(driver, 10)
attainment_tab = wait.until(EC.element_to_be_clickable((By.XPATH, "//p[text()='Attainment']")))
attainment_text = attainment_tab.text
isced_level_attainment_section = driver.find_element(By.XPATH, "//p[text()='Isced level attainment']")
isced_level_attainment_section.click() # Locate and click the "Tertiary" link
isced_level_attainment_text = isced_level_attainment_section.text
tertiary_link = wait.until(EC.element_to_be_clickable((By.XPATH, "(//div[@class='jss12']/div)[3]")))
tertiary_link.click()
teritery_link_text = tertiary_link.text
print(f"Subcategory Name: {subcategory_name}attaiment: {attainment_text}isced_level_attainment_text: {isced_level_attainment_text}tertiary_link_text: {teritery_link_text}")
```

```python
scraped_data = []

try:
    # Locate all elements dynamically
    xpath = "(//div[contains(@class, 'MuiCollapse-wrapperInner')])[3]/li"
    elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath)))

    # Loop through all elements using their index
    for index in range(1, len(elements)  + 1):
        try:
            # Construct dynamic XPath for each element
            link_xpath = f"({xpath})[{index}]"
            
            # Wait for and locate the clickable link
            link = wait.until(EC.element_to_be_clickable((By.XPATH, link_xpath)))
            driver.execute_script("arguments[0].scrollIntoView(true);", link)  # Optional scrolling
            time.sleep(4)  # Optional delay
            link.click()
            try:
               indiicator = wait.until(EC.presence_of_element_located((By.XPATH,"//div[@class='data-origin-title']"))).text
            except:
               indiicator = "No indicator found"
            try:
                description =  wait.until(EC.presence_of_element_located((By.XPATH, "//div[@class='info-card-content']/div/div"))).text  # Replace with the actual XPath
            except:
                description = "Description not found"
            
            try:
              technical_name =  wait.until(EC.presence_of_element_located((By.XPATH, "//div[@class='info-card-content']/div"))).text  # Replace with the actual XPath
            except:
               technical_name = "Technical name not found"
            try:
              Scale =  wait.until(EC.presence_of_element_located((By.XPATH, "(//div[@class='info-card-content']/div)[2]"))).text 
            except:
              Scale = "Not found"
            try:
    #
               related_indicators = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//div[@class='related-links']/a")))
    
    
               links_related_indicator = [link.get_attribute("href") for link in related_indicators]
            except Exception as e:
                print("No related indicators")
                print(f"Error: {e}")
            
            scraped_data.append({ 
                   "SubCatagory name": subcategory_name,
                  " Attaiment":isced_level_attainment_text,
                   "Sub-Sub category": teritery_link_text,
                   "Indicator": indiicator,
                   "description": description,
                   "technical_name": technical_name,
                   "Scale_name": Scale,
                   "Related Indicator":links_related_indicator

            })
   
        
            driver.back()
            
            time.sleep(4)

            elements = wait.until(EC.presence_of_all_elements_located((By.XPATH, xpath)))
            
            
            
            
        except Exception as e:
            print(f"Error processing element {index}: {e}")

except Exception as e:
    print(f"An error occurred: {e}")

    for data in scraped_data:
      print(data)


    

# Print the scraped data
```
